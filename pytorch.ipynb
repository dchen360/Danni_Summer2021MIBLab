```python
import torch
import os
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from torchvision.datasets import ImageFolder
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader
```

```python
# Load the data into train_dataloader and test_dataloader using ImageFolder
# Automatically assign the data with labels that correspond to the folder names
# ImageFolder returns dataset object which contains tuples => (image, class)

train_dataloader = ImageFolder(root=os.path.join(r'C:\Users\danni\OneDrive\Mini Project\Garbage Classification Dataset\MetalPaper_Train'))
test_dataloader = ImageFolder(root=os.path.join(r'C:\Users\danni\OneDrive\Mini Project\Garbage Classification Dataset\MetalPaper_Test'))

print(train_dataloader.classes) 

for x,y in train_dataloader:
    print(x)
    print(y)
```

```python
# Run this to test the dataloaders

# Get the 100th image from the train_dataloader
idx = 431
train_image, class_label = train_dataloader.__getitem__(idx)
plt.figure()
plt.imshow(train_image)
plt.title(f'Image at index {idx}. Class Label {class_label}')
plt.show() 

# Get the 5th image from the test_dataloader
idx = 5
test_image, class_label = test_dataloader.__getitem__(idx)
plt.figure()
plt.imshow(test_image)
plt.title(f'Image at index {idx}. Class Label {class_label}')
plt.show() 
```
